```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
class CustomDataset(Dataset):

  def __init__(self, features, labels):

    self.features = features
    self.labels = labels

  def __len__(self):
    return len(self.features)

  def __getitem__(self, idx):
        # Return one sample and its label
        x = self.features[idx]
        y = self.labels[idx]
        return x, y
```

```python
train_dataset = CustomDataset(X_train_tensor,y_train_tensor)
test_dataset = CustomDataset(X_test_tensor,y_test_tensor)
```

```python
train_dataset[10]
```

```python
train_loader = DataLoader(train_dataset,batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset,batch_size=32, shuffle=True)
```

```python
class MySimpleNN():

  def __init__(self,X):

    self.weights = torch.rand(X.shape[1],1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def farward(self,X):
    z = torch.matmul(X,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    loss = -(y_train_tensor * torch.log(y_pred)+ (1-y_train_tensor) * torch.log(1-y_pred)).mean()
    return loss
```

```python
import torch
import torch.nn as nn
class MyPyTorchNN(nn.Module):

  def __init__(self,num_features):
    super().__init__()

    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid

  def farward(self,X):
    z = self.linear(X)
    y_pred = self.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    loss = -(y_train_tensor * torch.log(y_pred)+ (1-y_train_tensor) * torch.log(1-y_pred)).mean()
    return loss
```

```python
import torch
import torch.nn as nn
class MyPyTorchMultiLayerNN(nn.Module):

  def __init__(self,num_features):
    super().__init__()

    self.linear1 = nn.Linear(num_features,3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3,1)
    self.sigmoid = nn.Sigmoid()


  def forward(self,X):
    z1 = self.linear1(X)
    h1 = self.relu(z1)
    z2 = self.linear2(h1)
    y_pred = self.sigmoid(z2)
    return y_pred

  def loss_function(self, y_pred, y):
    # clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    loss = -(y_train_tensor * torch.log(y_pred)+ (1-y_train_tensor) * torch.log(1-y_pred)).mean()
    return loss
```

```python
import torch
import torch.nn as nn
class MyPyTorchMultiLayerSequentialNN(nn.Module):

  def __init__(self,num_features):
    super().__init__()
    self.network = nn.Sequential(
      nn.Linear(num_features,3),
      nn.ReLU(),
      nn.Linear(3,1),
      nn.Sigmoid()
    )


  def forward(self,X):
    y_pred = self.network(X)
    return y_pred

  def loss_function(self, y_pred, y):
    # clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    loss = -(y_train_tensor * torch.log(y_pred)+ (1-y_train_tensor) * torch.log(1-y_pred)).mean()
    return loss
```

```python
import torch
import torch.nn as nn
class MyImprovedNN(nn.Module):

  def __init__(self,num_features):
    super().__init__()
    self.network = nn.Sequential(
      nn.Linear(num_features,1),
      nn.Sigmoid()
    )

  def forward(self,X):
    y_pred = self.network(X)
    return y_pred

loss_function = nn.BCELoss()
```

```python
learning_rate = 0.1
epochs = 25
```

```python
## training pipeline with data loader

# create model
model = MyImprovedNN(X_train_tensor.shape[1])

optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)
# loop
for epoch in range(epochs):

  for batch_features, batch_labels in train_loader:
    # farward pass
    y_pred = model(batch_features.float())
    # loss
    loss = loss_function(y_pred,batch_labels.reshape(-1,1).float())
    # clear gradient
    optimizer.zero_grad()

    # backward pass
    loss.backward()
    # update parameters
    # update parameters manually
    optimizer.step()

  #print loss in each epoch
  print(f"Epoch : {epoch+1}, Loss : {loss.item()}")
```

```python
# Model Evaluation using test loader
model.eval() # set model in evaluation mode
accuracy_list = []

with torch.no_grad():
  for batch_features, batch_labels in test_loader:

    # forward pass
    y_pred = model(batch_features)
    y_pred = (y_pred > 0.9).float() # convert probabilities to binary prediction

    # calculation of accuracy for the current batch
    batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
    accuracy_list.append(batch_accuracy)

# calculate overall accuracy
overall_accuracy = sum(accuracy_list)/len(accuracy_list)
print(f"Accuracy : {overall_accuracy:.4f}")

```
